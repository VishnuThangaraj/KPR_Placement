```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Clean a messy HR Dataset: fix missing values, rename columns, group by department to compute average salary, merge with location table


Load Dataset


```python
df = pd.read_csv('Dataset/employeeDataset.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

Check and Rename Columns


```python
df.columns
```

```python
df.rename(columns={'Email':'Email_ID'}, inplace=True)
df.columns
```

Change DataTypes


```python
df['Age'] = df['Age'].astype('Int8')
df['Phone'] = df['Phone'].astype(str)

df[['Age', 'Phone']].head()
```

```python
df.dtypes
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
df.head()
```

Handle Missing Values


```python
# Phone
df['Phone'] = df['Phone'].str.slice(1)
df['Phone'].head()
```

```python
mean_age = df['Age'].mean().astype(int)
df['Age'] = df['Age'].fillna(mean_age)
df.head()
```

Split Department and Region


```python
df[['Department', 'Region']] = df['Department_Region'].str.split('-', n=1, expand=True)
df.head()
```

Drop Department_Region


```python
df.drop(columns=['Department_Region'], inplace=True)
df.head()
```

Group by Department


```python
dept_count = df.groupby('Department')['Employee_ID'].count()
dept_count.head()
```

```python
avg_salary = df.groupby('Department')['Salary'].mean().map('{:.2f}'.format)
avg_salary
```
