## Slicing


```python
arr = np.random.randint(1,10,5)
print('Array :',arr)
print('Elements from 2nd Index :',arr[2:])
print('Elements from 2nd - 3th Index :',arr[2:4])
print('Elements from 2nd - 3th Index :',arr[2:4])
print('Alternate Elements :',arr[::2])
print('Index Selection :', arr[[0,2,4]])
```

## Mathematical Operations


```python
arr = np.random.randint(1,10,5) 
print('Array :',arr)
print('Addition :',arr+10) # Scalar Broadcasting
```

```python
arr1 = np.random.randint(1,10,size=(3,4))
print('Array 1 :\n',arr1,"\n")

arr2 = np.random.randint(1,10,size=(3,4))
print('Array 2 :\n',arr2,"\n")

print('Addition :\n',arr1+arr2,"\n") # Row broadcasting
print('Subtraction :\n',arr1-arr2,"\n") # Row broadcasting
```

1. Create 2D NumPy array of student scores; compute class average, highest scorer, normalize scores to 0-1 range


```python
score = np.random.randint(1,101,size=(10,5))
print('Scores :\n',score,'\n')

student_total_score = np.sum(score, axis=1)
print('Student Total Score : \n',student_total_score,'\n')
```

```python
print('Highest Score :', np.max(student_total_score))
print('Lowest Score :', np.min(student_total_score))
print('Class Average Score :', np.mean(student_total_score))
```

```python
print('Student Highest Score :', np.max(score, axis=1))
print('Student Lowest Score :', np.min(score, axis=1))
print('Student Average Score :', np.mean(score, axis=1))
```

```python
print('Subject Highest Score :', np.max(score, axis=0))
print('Subject Lowest Score :', np.min(score, axis=0))
print('Subject Average Score :', np.mean(score, axis=0))
```

```python
normalized_score = (score - np.min(score)) / (np.max(score) - np.min(score))
print('Normalized Score :\n',normalized_score,'\n')
```
