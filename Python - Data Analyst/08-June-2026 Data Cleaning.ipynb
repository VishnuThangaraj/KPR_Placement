# Machine Learning

### Why was Machine Learning Created?
* Before Machine Learning, software relied entirely on manually written rules created by programmers. These rules worked well for simple and predictable tasks but became difficult to manage when dealing with complex real-world problems.

### Identifying Cats in Images
Traditional Programming:
* IF Animal has whiskers
* AND Animal has pointed ears
* AND Animal has four legs
* THEN Cat

# Problems:
* Every cat looks different.
* Some dogs have similar features.
* Cats can appear in different positions, sizes, and lighting conditions.
* Writing rules for every possible situation is nearly impossible.
* The number of rules becomes very large and difficult to maintain.
* Traditional programs often fail when they encounter new or unexpected examples.

### Types of Machine Learning

Machine Learning Algorithms are mainly classified into four types based on the availability of labeled data and the learning process.
* Supervised Learning
* Unsupervised Learning
* Semi-Supervised Learning
* Reinforcement Learning

### 1. Supervised Learning

Supervised Learning is a type of Machine Learning where the model is trained using Labeled Data.

Labeled Data=Input + Correct Output

The Algorithm learns the relationship between input variables (features) and output variables (target).

**Example:**

Student Dataset

<div style="zoom:140%;">

| Study Hours | Marks |
|-------------|-------|
| 2 | 40 |
| 4 | 60 |
| 6 | 80 | 

</div>

**The Model Learns:**
* More Study Hours -> Higher Marks

**Now if a Student Studies 5 Hours:**

**Real-Life Applications**

* Student Performance Prediction
* Spam Email Detection

### 2. Unsupervised Learning

* Unsupervised Learning is a type of Machine Learning where the model learns from Unlabelled Data.
* Only inputs are available, no correct answers are provided.

**Example:**

|----------|-----|--------|
| A | 25 | 30000 |
| B | 30 | 45000 |

Possible Output:

* Group 1 -> Young Customers
* Group 2 -> High-Income Customers

**Main Goal**
* Find hidden structures and patterns in data.

**Applications:**
* Customer Segmentation
* Market Basket Analysis
* Fraud Detection
* Recommendation Systems
* Social Network Analysis

**Example:**
* Customers with similar buying habits

Algorithm:

* K-Means Clustering

**B.Association**
* Finding Relationship Between Items.

**Example:**
* People Who buy Milk also buy bread.
* used in supermarkets.

**Advantages**
* No Labeled Data Required.
* Helps Discover Hidden Patterns

**Disadvantages:**
* Harder to Evaluate

### 3. Semi-Supervised Learning
* Semi-Supervised Learning Combines Both:
* Small Amount of Labeled Data + Large Amount of Unlabelled Data.

**Why Use it?**
* Labeling Data is expensive and Time Consuming.

**Example:**
* Hospital Has:
   * 1,000 X-ray images labeled by doctors (e.g., Normal or Pneumonia)
   * 50,000 X-ray images without labels

Instead of Labelling all images, the model learns from both labelled and lunlabelled data.

**Example:**

Suppose we have:
| Image | Label |
|-------|-------|
| Cat Image | Cat |
| Dog Image | Dog |
| Thousands of Images | No Label |


**Applications:**
* Medical Image Analysis
* Speech Recognition
* Face Recognition
* Text Classification

**Advantages:**
* Requires fewer labeled records
* Better Performance than purely unsupervised learning
* Reduces labeling cost.

**Disadvantages:**
* More Complex to Implement
* Incorrect Labels can affect Performance.

* The goal is to Maximize Rewards.

**Example: Self-Driving Car**

**Actions:**
* Turn Left
* Turn Right
* Brake
* Accelerate

**Applications:**
* Self-Driving Cars
* Robotics
* Game Playing AI
* Route Optimization
* Industrial Automation

**Advantages:**
* Learns from Experience
* Handles dynamic environments.
* Can achieve very high performance.

**Disadvantages:**
* Requires many training iterations.
* Computationally Expensive.
* Complex to design reward systems.

# Overfitting & Underfitting

## Overfitting

**Overfitting** occurs when a machine learning model learns the training data too closely, including its noise and random variations. As a result, the model performs very well on the training data but poorly on unseen test data.

### Characteristics

- Very high training accuracy
- Low test accuracy
- Low bias
- High variance
- Poor generalization

### Detection Using Learning Curves

- Training score remains very high.
- Validation score is much lower.
- A large gap exists between the training and validation curves.

## Underfitting

**Underfitting** occurs when a machine learning model is too simple to learn the underlying patterns in the data. Consequently, it performs poorly on both training and test datasets.

### Characteristics

- Low training accuracy
- Low test accuracy
- High bias
- Low variance
- Poor learning of data patterns

### Detection Using Learning Curves

- Both training and validation scores are low.
- The curves remain close to each other.
- Performance does not improve significantly.

---

# Bias vs Variance Tradeoff

The **Bias-Variance Tradeoff** is the balance between a model's ability to learn patterns from training data (**bias**) and its ability to generalize to unseen data (**variance**).

The objective is to find a model complexity that minimizes both bias and variance, resulting in the best predictive performance.

---

## Bias

### Definition

Bias is the error introduced when a model makes overly simplistic assumptions about the data.

### High Bias

- Model is too simple.
- Important patterns are missed.
- Causes underfitting.

### Example

Using **Linear Regression** to model a highly nonlinear relationship.

> **High Bias → Underfitting**

---

## Variance

### Definition

Variance is the error caused when a model becomes too sensitive to the training data and learns noise instead of general patterns.

### High Variance

- Model is too complex.
- Memorizes training data.
- Causes overfitting.

### Example

A very deep **Decision Tree** fitting every training sample perfectly.

> **High Variance → Overfitting**

---

## The Sweet Spot

### Definition

The **Sweet Spot** is the optimal balance between bias and variance where the model captures meaningful patterns from the data without memorizing noise.

### Characteristics

- Low training error
- Low validation error
- Good generalization
- Best predictive performance
