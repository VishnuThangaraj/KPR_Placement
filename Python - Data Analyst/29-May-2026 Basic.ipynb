## Perfect Number

```python
# Perfect Number
num = int(input("Enter a number: "))
sum_val = 0

for i in range(1, num):
    if num % i == 0:
        sum_val += i
        
print(f"{'Perfect Number' if sum_val == num else 'Not a Perfect Number'}")
```

## Neon Number

```python
# Neon Number
num = int(input("Enter a number: "))
num_sqr, total = num*num, 0


while num_sqr>0:
    total += num_sqr%10
    num_sqr = num_sqr // 10
    
print(f"{'Neon Number' if num == total else 'Not a Neon Number'}")
```

## Armstrong Number

```python
# Armstrong Number
num = int(input("Enter a number: "))
digit_count, num_temp = 0, num
total = 0

while num_temp > 0:
    digit_count += 1
    num_temp = num_temp//10

num_temp = num

while num_temp > 0:
    digit = num_temp%10
    total += digit ** digit_count
    num_temp = num_temp//10 

print(f"{'Armstrong Number' if num == total else 'Not an Armstrong Number'}")
```

## Strong Number

```python
# Strong Number
num = int(input("Enter a number: "))
fact_sum = 0

while num > 0:
    temp = 0
    for i in range(1, num+1):
        temp *= i
    num = num // 10
    fact_sum += temp
    
print(f"{'Strong Number' if num == fact_sum else 'Not a Strong Number'}")
```
