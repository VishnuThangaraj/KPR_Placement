```python
import numpy as np
import pandas as pd
```

# SERIES


```python
arr = pd.Series(np.arange(1,5))
print(f'Array :\n{arr}')
```

```python
print(f'Array Values : {arr.values}\n')
print(f'Array Index : {arr.index}\n')
print(f'Array Data Type : {arr.dtype}\n')
```

```python
arr = pd.Series(['Thilak', 'Siva', 'Vignesh', 'Vishnu'])
print(f'Array : \n{arr}\n')
print('Value in 1st Index :',arr[1])
```

```python
print(f'Iloc : {arr.iloc[1]}\n')
```

## Slicing


```python
print(f'loc : \n{arr.loc[1:3]}\n')
print(f'Iloc : \n{arr.iloc[1:3]}\n')
```

```python
print(f'Iloc : \n{arr.iloc[0:3]}\n')
```

## Indexing


```python
arr = pd.Series([1,2,3,4,5], index=['A','B','C','D','E'])
print(f'Array : \n{arr}\n')
print(f'Indexing (B): {arr["B"]}\n')
```

## Arithmetic Operations


```python
print(f'Array Addition: \n{arr+5}\n')
print(f'Array Subtraction: \n{arr-5}\n')
```

## Statistical Operations


```python
print(f'Sum : {arr.sum()}')
print(f'Min : {arr.min()}')
print(f'Max : {arr.max()}')
print(f'Max : {arr.max()}')
print(f'Mean : {arr.mean()}')
print(f'Median : {arr.median()}')
print(f'Standard Deviation : {arr.std()}')
```

## String Operations


```python
name = pd.Series(['Thilak', 'Siva', 'Vignesh', 'Vishnu', 'Subash'])
print(f'Array: \n{name}\n')
```

```python
print(f'String Upper: \n{name.str.upper()}\n')
print(f'String Lower: \n{name.str.lower()}\n')
print(f'String Length: \n{name.str.len()}\n')
```

```python
arr = pd.Series({'Vishnu':23, 'thilak': 52, 'Vignesh': 33})
print(f'Array: \n{arr}\n')
```

```python
print(f'Value of Vishnu : {arr["Vishnu"]}')
```

```python
arr1 = pd.Series([10,10,30], index=['A','B','B'])
print(f'Array 1: \n{arr1}\n')

arr2 = pd.Series([1,2,3], index=['B','C','D'])
print(f'Array 2: \n{arr2}\n')

# Addition
print(f'Array Addition - Index Based: \n{arr1+arr2}\n')
```

```python
print(f'Elements Count: {arr1.count()}')
print(f'Unique Elements: {arr1.unique()}')
print(f'Unique Elements Count: {arr1.nunique()}')
```

# DATA FRAME


```python
data = pd.DataFrame({
    'name':['Vishnu', 'Vignesh', 'Thilak'],
    'mark': [34, 65, 99],
    'age':[12, 20, 25]
})
print(f'Data Frame :\n',data)
```

```python
roll_no = pd.Series([11,12,49,53, 55])
name = pd.Series(['Vishnu', 'Vignesh', 'Thilak', 'Siva', 'Subash'])
age = pd.Series([21,22,24,23,19])
attendance = pd.Series([True, False, False, True, True])

print(f'Roll No: {roll_no}\n')
print(f'Name: {name}\n')
print(f'Age: {age}\n')
print(f'Attendance: {attendance}\n')
```

Create DataFrame with Series


```python
df = pd.DataFrame({
    'roll_no': roll_no,
    'name':name,
    'age': age,
    'attendance': attendance
})
print(f'Data Frame:\n{df}')
```

Convert 2D List to DataFrame


```python
data = [
    ['Vishnu', 22, 89],
    ['Vignesh', 21, 93],
    ['Siva', 20, 78],
]
print(f'Data : {data}\n')

df = pd.DataFrame(data)
print(f'DataFrame: \n{df}\n')

df = pd.DataFrame(data, columns=['Name','Age','Mark'])
print(f'DataFrame with Columns: \n{df}')
```

Convert Dictionary to DataFrame


```python
data = [
    {'Name': 'Vishnu', 'Age': 22, 'Mark': 89},
    {'Name': 'Vignesh', 'Age': 21, 'Mark': 93},
    {'Name': 'Siva', 'Age': 20, 'Mark': 78},
]

print(f'Data : {data}\n')

df = pd.DataFrame(data)
df
```

Add New Row


```python
df = pd.concat([df, pd.DataFrame([{'Name': 'Subash', 'Age': 29, 'Mark': 99}])], ignore_index=True)
print(f'DataFrame: \n{df}\n')
```

Add New Column


```python
df['Grade'] = ['A', 'A+', 'B', 'C']
df
```

Accessing Column


```python
df[['Name','Age']]
```

Accessing Row


```python
df.iloc[1]
```

```python
print(f'Shape : {df.shape}')
print(f'No of Rows : {df.shape[0]}') 
print(f'No of Columns : {df.shape[1]}') 
```

```python
print(f'Data Types : \n{df.dtypes}')
```

```python
print(f'Data Frame Info : \n{df.info()}')
```

```python
print(f'Data Frame Describe : \n{df.describe()}')
```

```python
df.loc[0,'Grade'] = 'C+'
df
```
