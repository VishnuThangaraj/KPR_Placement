## PIVOT TABLE


```python
df = pd.read_csv('Dataset/people.csv')
df.head()
```

```python
df.columns
```

Create Age Column


```python
df['Date of birth'] = pd.to_datetime(df['Date of birth'])
df.head()
```

```python
today = pd.Timestamp.today()
df['Age'] = (
    (today - df['Date of birth']).dt.days // 365
)
df.head()
```

```python
pivot1 = pd.pivot_table(df, values='Age', index='Sex')
pivot1
```

Aggregate Function in pivot table


```python
pivot2 = pd.pivot_table(df, values='Age', index='Sex', aggfunc='count')
pivot2
```

Multiple Aggregation


```python
pivot3 = pd.pivot_table(df, values='Age', index='Sex', aggfunc=['count', 'mean', 'min', 'max'])
pivot3
```

Pivot Table with Margins


```python
pivot4 = pd.pivot_table(df, values='Age', index='Sex', aggfunc='mean', margins=True)
pivot4
```

Group by Job Title and Gender


```python
pivot5 = pd.pivot_table(df, values='Age', index='Job Title', columns='Sex', aggfunc='count')
pivot5.head()
```

Count People by Job and Gender, Fix NaN using fill_values


```python
pivot6 = pd.pivot_table(df, values='User Id', index='Job Title', columns='Sex', aggfunc='count', fill_value=0)
pivot6.head()
```
